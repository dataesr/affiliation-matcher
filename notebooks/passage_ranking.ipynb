```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv

load_dotenv()

ODS_API_KEY = os.getenv("ODS_API_KEY")
ODS_PAYSAGE_KEY = os.getenv("ODS_PAYSAGE_KEY")

SCANR_DUMP_URL = 'https://scanr-data.s3.gra.io.cloud.ovh.net/production/organizations.jsonl.gz'
```

```python
df = pd.read_csv(
    f"https://data.enseignementsup-recherche.gouv.fr/explore/dataset/{ODS_PAYSAGE_KEY}/download/?format=csv&apikey={ODS_API_KEY}",
    sep=";",
    low_memory=False,
)
```

```python
def download_data() -> list:
    if 'jsonl' in SCANR_DUMP_URL:
        data = pd.read_json(SCANR_DUMP_URL, lines=True)
    else:
        r = requests.get(SCANR_DUMP_URL)
        data = pd.DataFrame(r.json())
    return data
```

```python
data = download_data()
```
